# taxa_abundance_plot

```python
help(api.taxa_abundance_plot)
```

Below is the simplest example.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', figsize=(14, 7), show_legend=True)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-1.png')
```

We can increase the taxonomic rank to species by setting `level=7`.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', figsize=(14, 7), level=7)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-2.png')
```

We can only display the top eight most abundant species with `count=8`.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', figsize=(14, 7), level=7, count=8, show_legend=True, legend_loc='upper right', legend_short=True)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-3.png')
```

To sort the samples by body-site, use `by=['body-site']`.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', by=['body-site'], figsize=(14, 7), level=7, count=8, show_legend=True, legend_loc='upper right', legend_short=True)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-4.png')
```

We can replace the x-axis labels with the `label_columns` option.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', label_columns=['body-site'], by=['body-site'], width=0.9, figsize=(14, 7), level=7, count=8, show_legend=True, legend_loc='upper right', legend_short=True)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-5.png')
```

We can only display the N and T samples with `exclude_samples={'body-site': ['gut', 'tongue']}`.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', exclude_samples={'body-site': ['gut', 'tongue']}, label_columns=['body-site'], by=['body-site'], width=0.9, figsize=(14, 7), level=7, count=8, show_legend=True, legend_loc='upper right', legend_short=True)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-6.png')
```

We can also control the bar colors with the `colors` option.

```python
api.taxa_abundance_plot('taxa-bar-plots.qzv', colors=['tab:blue', 'tab:orange', 'tab:gray'], exclude_samples={'body-site': ['gut', 'tongue']}, label_columns=['body-site'], by=['body-site'], width=0.9, figsize=(14, 7), level=7, count=3, show_legend=True, legend_loc='upper right', legend_short=True)
plt.tight_layout()
#plt.savefig('taxa-bar-plots-7.png')
```
